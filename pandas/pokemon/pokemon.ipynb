```python
import pandas as pd
```

```python
data_frame = pd.read_csv("pokemon.csv")
data_frame.head()
```

```python
data_frame.info()
```

Find all pokemons witn an Atack value greater than 150

```python
(data_frame["Attack"] > 150).value_counts() # Count it
data_frame.loc[data_frame["Attack"] > 150] # loc method
data_frame.query("Attack > 150").head(3) # query method
```

Select all pokemons with Speed of 10 or less

```python
(data_frame["Speed"] <= 10).sum() # Count it
data_frame.query("Speed <= 10")
```

How many Pokemons have a Sp. Def value of 25 or less?

```python
(data_frame["Sp. Def"] <= 25).sum() # Way 1
(data_frame["Sp. Def"] <= 25).value_counts() # Way 2
```

Select all the legendary pokemons

```python
data_frame.loc[data_frame["Legendary"] == False]
```

Find the pokemon outlier (with tremendous defense but low attack)

```python
data_frame.sort_values(by=["Defense", "Attack"], ascending=[False, True]).head(1) # Order by Defense and attack but Defense in ascending mode and Attack not 
```

How many pokemons are of 'Type 1' Fire and 'Type 2' flying?

```python
(
  (data_frame["Type 1"] == "Fire") &
  (data_frame["Type 2"] == "Flying")
).sum()
```

How many pokemons exist that are of type Poison in either Type 1 or Type 2

```python
(
  (data_frame["Type 1"] == "Poison") |
  (data_frame["Type 2"] == "Poison")
).sum()
```

What pokemon of type 1 Ice has the stronguest defence?

```python
data_frame.sort_values(by="Defense", ascending=False).loc[data_frame["Type 1"] == "Ice"].head(1)
```

What is the most common type of legendary pokemons?

```python
data_frame.loc[data_frame["Legendary"] == True, "Type 1"].value_counts()
```

What is the most powerful pokemon from the firt 3 generation, of type 1 of water and fire

```python
data_frame.sort_values(by="Attack", ascending=False).loc[(data_frame["Generation"] <= 3) & (data_frame["Type 1"].isin(["Water", "Fire"]))].head(2)
```

What is the most powerful by Total that is Type 1 and 2 Dragon and from the last two generations

```python
data_frame["Generation"].max() # 6. I have to looking for greater or equal than 5
data_frame.sort_values(by="Total", ascending=False).loc[
  (data_frame["Generation"] >= 5) &
  ((data_frame["Type 1"] == "Dragon") |
  (data_frame["Type 2"] == "Dragon"))
].head(1)
```

Select those pokemons that are Type 1 Water and Type 2 Flying. Store the selection in the variable water_flying_df

```python
water_flying_df = data_frame.loc[(data_frame["Type 1"] == "Water") & (data_frame["Type 2"] == "Flying")]
```

Perform a selection in your dataframe of all Legendary pokemons that are of type 1 Fire. Select only the columns Name, Attack and Generartion. Store the result in a variable called legendary_fire_df

```python
legendary_fire_df = data_frame.loc[(data_frame["Type 1"] == "Fire") & (data_frame["Legendary"] == True), ["Name", "Attack", "Generation"]] # df.loc[(CONDITIONS FOR INDEX), (COLUMNS)]
legendary_fire_df
```

Select the pokemons that are below 5% speed or 95% above speed.

```python
bottom_5 = data_frame["Speed"].quantile(.05) # 5%
top_5 = data_frame["Speed"].quantile(.95) # 95%

data_frame.loc[(data_frame["Speed"] > top_5) | (data_frame["Speed"] < bottom_5)]
```
