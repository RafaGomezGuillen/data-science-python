# Seaborn

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style = "darkgrid")

data_frame = sns.load_dataset("mpg")

data_frame.head(2)
```

## Relational Charts
### Dot Chart
`sns.relplot(x, y, data_set)`

```python
# height = 5. Height Chart (not very important)
# style = "cylinders". Dot style
# size = "cylinders". Dot size 
# sizes = (15, 20). Dot size values (min, max)
# hue = origin.Another dimension can be added to the chart by coloring the dots according to a third variable (hue).
sns.relplot(x = "weight", y = "horsepower", data = data_frame, height = 5, hue = "origin", style = "cylinders", size = "cylinders", sizes = (15, 200))
```

### Line Chart
`sns.relplot(x, y, data_set, kind = "line")`

```python
sns.relplot(x = "weight", y = "horsepower", data = data_frame, hue = "origin", kind = "line")
```

```python
dates = pd.DataFrame(dict(time = pd.date_range("2020-01-01", periods = 200), value = np.random.randn(200).cumsum()))

sns.relplot(x = "time", y = "value", kind = "line", data = dates, height = 7)
```

## Categorical Charts

```python
data_frame = sns.load_dataset("titanic")
data_frame.head(2)
```

### Count Plot

```python
# A count plot is like a bar plot, but the estimator is counting  the number of occurances
sns.countplot(x ="sex", data = data_frame)
```

### Catplot
`sns.catplot(x, y, data_set)`

```python
sns.catplot(x = "pclass", y = "age", data = data_frame)
```

If we want to prevent the dots from overlapping, we use the `kind = 'var'`. We can add a hue as well

```python
sns.catplot(x = "pclass", y = "age", data = data_frame, kind = "violin", hue = "sex")
```

### Strip plot

```python
# Jitter spreads data points out so that they aren't stacked on top of each other
# Hue breaks data into men and women
# Dodge separates the men and women data
sns.stripplot(x = "pclass", y = "age", data = data_frame, hue = "sex", jitter = True, dodge = True)

plt.legend(loc=0)
```

### Boxplot, set_context(), sns.color_palette() (styles)
`sns.boxplot(x, y, data_set)`

```python
fig, ax = plt.subplots(figsize = (15, 6))

sns.set_context('notebook', font_scale = 1.4, rc = {"lines.linewidth": 3.5})
# ax. Equivalent to height
sns.boxplot(x = 'pclass', y = 'age', data = data_frame, hue = "sex", ax = ax, palette = sns.color_palette("husl", 8))
```

We can modify the orientation of the chart `orient = "var"`

```python
fig, ax = plt.subplots(figsize = (15, 6))

sns.boxplot(x = 'age', y = 'pclass', data = data_frame, hue = "sex", orient = "h")
```

### Bar plots
`sns.catplot(x, y, data_set, kind = 'bar')`

```python
# estimator = np.count_nonzero. It will count the number of non-null observations for each combination of the categories
sns.catplot(x = 'pclass', y = 'survived', hue = 'sex', data = data_frame, kind = 'bar', estimator = np.count_nonzero)
```

## Charts for Distributions
### Histplot

```python
fig, ax = plt.subplots(figsize = (20, 5))
# kde = True. Show the line (optional)
# bins = 10. Number of bins (optional)
# element = "step". Style element histogram (optional)
# multiple = "stack". Determines how multiple elements should be treated with a HUE (optional)
sns.histplot(x = "age", data = data_frame, kde = True, ax = ax, bins = 10, hue = "sex", element = "step", multiple = "stack")
```

```python
sns.histplot(x = "age", data = data_frame, kde = True, hue = "sex", stat = "probability")
```

To use the parameter `hue` in histograms we use `sns.displot()`

```python
sns.displot(x = "age", data = data_frame, col = "sex", kind = "hist")
```

### Distribution Plot

```python
crash_df = sns.load_dataset('car_crashes')

# kde=False removes the KDE
# Bins define how many buckets to divide the data up into between intervals
# For example put all profits between $10 and $20 in this bucke
sns.displot(crash_df["not_distracted"], bins = 25)
```

## Regression charts
`sns.regplot(x, y, data_set)`

```python
data_frame = sns.load_dataset("tips")
data_frame.head(2)
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
# line_kws={'color':'red'}. Color line
# color = "green". Dot color
sns.regplot(x = 'tip', y = 'total_bill', data = data_frame, ax = ax, line_kws={'color':'red'}, color = 'green')
```

In addition to regplot we have a very similar method, which the lmplot, the operation is practically identical. 
In lmplot we can use the `hue` parameter.
`sns.lmplot(x, y, data_set)`

```python
sns.lmplot(x = 'tip', y= 'total_bill', data = data_frame, height = 6, hue = 'smoker', ci = False)
```

If we put everything together a bit, we could end up with a graph of the following type

```python
sns.lmplot(x = 'tip', y = 'total_bill', hue = 'smoker', data = data_frame, col = 'sex')
```

### Jointplot
This chart combines regression with distribution.
`sns.jointplot(x, y, data_set)`

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = data_frame, kind = 'reg')
```

```python
# Jointplot compares 2 distributions and plots a scatter plot by default
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind="reg")
# sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind="kde")
# sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind="hex")
```

### KDE Plot

```python
# Get just the KDE plot
sns.kdeplot(crash_df["alcohol"])
```

## Multiple Charts

```python
data_frame = sns.load_dataset("titanic")
data_frame.head(2)
```

### FacetGrid
`chart = sns.FacetGrid(data_set, col = 'value')`
To complete the chart: `chart.map(chart_type, value)`

```python
chart = sns.FacetGrid(data = data_frame, col = 'sex')
chart.map(sns.histplot, 'age')
```

```python
chart = sns.FacetGrid(data = data_frame, col = 'pclass', hue = 'sex')
chart.map(sns.scatterplot, 'age', 'fare')
chart.add_legend()
```

```python
chart = sns.FacetGrid(data = data_frame, col = 'pclass')
# s = 60. Dot size
chart.map(sns.scatterplot, 'age', 'fare', color = 'green', s = 60)

for ax in chart.axes.flat:
    ax.axhline(y = 300) # line

# ticks
# xticks 0 to 75
# yticks 0 400
chart.set(xlim = (0, 75), ylim = (0, 400))
```

## Matrix Plots
### Heatmaps

```python
# To create a heatmap with data you must have data set up as a MATRIX where variables
# are on the columns and rows
flights = sns.load_dataset("flights")
flights = flights.pivot_table(index = "month", columns = "year", values = "passengers")

# You can separate data with lines
sns.heatmap(flights, cmap = 'Blues', linecolor = 'white', linewidth = 1)
```

## Color Palettes

```python
data_frame = sns.load_dataset("tips")

plt.figure(figsize=(8,6))
sns.set_style('dark')
sns.set_context('talk')

# You can use Matplotlibs color maps for color styling
# https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html
sns.stripplot(x = 'day', y = 'total_bill', data = data_frame, hue = 'sex', palette = 'seismic')

# Add the optional legend with a location number (best: 0, upper right: 1, upper left: 2, lower left: 3, lower right: 4)
plt.legend(loc = 0)
```
